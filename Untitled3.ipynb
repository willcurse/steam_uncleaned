# 2.1 Handling Missing Values

```python
#drop rows where review number is missing 
data.dropna(subset=['Review_no'])
data.head()
```

```python
data.isnull().sum()
```

```python
#printing all the missing value column
data[data.isna().any(axis=1)]

```

```python
#changong the price datatype
data['Price'] = data['Price'].astype(str)  # Ensure all values are strings
data['Price'] = data['Price'].str.replace('$', '', regex=False).str.strip()
data['Price'] = data['Price'].replace('Free To Play', '0')  # Replace non-numeric values
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')  # Convert to numeric
```

```python
#filling missing values price with mean price
round(data['Price'].fillna(data['Price'].mean()))
```

```python
#forward/backward fill 
data['Release_date'].fillna(method='ffill')
```

# 2.2 Correcting dataTypes

```python
#type conversion of datetime string
data['Release_date']=pd.to_datetime(data['Release_date'],errors='coerce')
data.head()
```

```python
#type conversion of price
data['Price']=data['Price'].astype(str)
#replacing $ sign with empty string
data['Price']=data['Price'].str.replace('$','',regex='False').str.strip()
data
```

```python
#removing the user reciew from review no
data['Review_no']=data['Review_no'].str.replace('User Reviews','').str.strip()
data['Review_no']=data['Review_no'].str.replace(',','')
data.head()
```

# 2.3 -: Removing the duplicate values

```python
#check all the rows and remove the dupliacte values
duplicate_rows = data[data.duplicated()]
print(duplicate_rows)

#checking according to tge particular row
duplicate_names = data.duplicated(subset=['Review_no'])
# print(duplicate_names)

# Remove duplicate rows
data_cleaned = data.drop_duplicates(inplace=True)
print(data_cleaned)
```

# 2.4 Handling the outliers

```python
sns.boxplot(data['Price']) #with outliers
```

```python
#changong the price datatype
data['Price'] = data['Price'].astype(str) 
data['Price'] = data['Price'].str.replace('$', '', regex=False).str.strip()
data['Price'] = data['Price'].replace('Free To Play', '0')  
data['Price'] = pd.to_numeric(data['Price'], errors='coerce') 
price_cap = data['Price'].quantile(0.99)

data['Price'] = data['Price'].apply(lambda x: price_cap if x > price_cap else x)
```

```python
sns.boxplot(data['Price']) #after removing the outliers
```
